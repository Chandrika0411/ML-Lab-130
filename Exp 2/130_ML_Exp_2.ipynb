```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

## Step 1 :  Data Collection

```python
import seaborn as sns

# Load Titanic Dataset
titanic_data = sns.load_dataset('titanic')
```

## Step 2 : Basic Data Inspection

```python
print(titanic_data.shape)
```

```python
# Display first few records
titanic_data.head()
```

```python
# Display last few records
titanic_data.tail()
```

```python
# Review the structure of the dataset
titanic_data.info()
```

```python
titanic_data.describe()
```

```python
# Identifying missing values
missing_values = titanic_data.isnull().sum()
missing_values
```

```python
# Dealing with missing values
# Dropping colums with excessive missing data
new_titanic_df = titanic_data.drop(columns = ['deck'])
```

```python
# Imputing median age for missing age data
new_titanic_df['age'].fillna(new_titanic_df['age'].median(), inplace = True)
```

```python
# Identifying missing values
print(new_titanic_df.isnull().sum())
```

```python
data = new_titanic_df
data['embark_town'].dtype
```

```python
data['embark_town'].unique()
```

```python
data['embark_town'].fillna(data['embark_town'].mode()[0], inplace = True)
```

```python
# Identifying missing values
data.isnull().sum()
```

```python
data['embarked'].dtype
```

```python
data['embarked'].unique()
```

```python
data['embarked'].fillna(data['embarked'].mode()[0], inplace = True)
```

```python
# Identifying missing values
data.isnull().sum()
```

## Step 4: Encode Categorical Variables

```python
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['embarked'] = le.fit_transform(data['embarked'])
```

```python
# Display first few records
data.head()
```

```python
# Review the structure of the dataset
data.info()
```

## Step 5: Select Features and Target

```python
data = data[['pclass', 'sex', 'age', 'fare', 'embarked', 'survived']]

x = data[['pclass', 'sex', 'age', 'fare', 'embarked']]
y = data['survived']
```

## Step 6: Train - Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.3, random_state = 42)
```

## Step 7: Train model

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train, y_train)
```

## Step 8: Prediction

```python
y_pred = model.predict(X_test)
```

## Step 9: Checking Accuracy

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```
