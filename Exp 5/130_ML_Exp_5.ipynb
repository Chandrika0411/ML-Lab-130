## Experiment-5: Disease Prediction using Naive Bayes and Neural Network with Comparison of Classifiers.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score,roc_auc_score, f1_score, confusion_matrix
```

```python
# Load real dataset
data = load_breast_cancer()

X = data.data
Y = data.target

print("Classes:", data.target_names)    # malignant, benign

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 30)
```

## Part A: Training and Test Accuracy

### 1. Logistic Regression

```python
# Logistic Regression
lr = LogisticRegression(max_iter = 5000)
lr.fit(X_train, Y_train)
pred_lr = lr.predict(X_test)
print("Logistic Test Accuracy:", accuracy_score(Y_test, pred_lr))
pred_lr_train = lr.predict(X_train)
print("Logistic Train Accuracy:", accuracy_score(Y_train, pred_lr_train))
```

### 2. Decision Tree

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, Y_train)
pred_dt = dt.predict(X_test)
print("Decision Tree Test Accuracy:", accuracy_score(Y_test, pred_dt))
pred_dt_train = dt.predict(X_train)
print("Decision Tree Train Accuracy:", accuracy_score(Y_train, pred_dt_train))
```

### 3. K Nearest Neighbor (KNN)

```python
# K Nearest Neighbor
knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
pred_knn = knn.predict(X_test)
print("K Nearest Neighbor Test Accuracy:", accuracy_score(Y_test, pred_knn))
pred_knn_train = knn.predict(X_train)
print("K Nearest Neighbor Train Accuracy:", accuracy_score(Y_train, pred_knn_train))
```

### 4. Naive Bayes

```python
# Naive Bayes
nb = GaussianNB()
nb.fit(X_train, Y_train)
pred_nb = nb.predict(X_test)
print("Naive Bayes Test Accuracy:", accuracy_score(Y_test, pred_nb))
pred_nb_train = nb.predict(X_train)
print("Naive Bayes Train Accuracy:", accuracy_score(Y_train, pred_nb_train))
```

### 5. Neural Network

```python
# Neural Network
nn = MLPClassifier(max_iter = 2000)
nn.fit(X_train, Y_train)
pred_nn = nn.predict(X_test)
print("Neural Network Test Accuracy:", accuracy_score(Y_test, pred_nn))
pred_nn_train = nn.predict(X_train)
print("Neural Network Train Accuracy:", accuracy_score(Y_train, pred_nn_train))
```

## Part C: Confusion Matrix and Error Analysis

### 1. Logistic Regression

```python
# Example Confusion Matrix, Type-I and Type-II Error
confusion_matrix_lr = confusion_matrix(Y_test, pred_lr)
print("Logistic Regression Confusion Matrix:")
print(confusion_matrix_lr)

TN1, FP1, FN1, TP1 = confusion_matrix_lr.ravel()
type1_lr = FP1 / (FP1 + TN1)    # False Positive Rate
type2_lr = FN1 / (FN1 + TP1)    # False Negative Rate

print("FP (Type-I):", FP1)
print("FN (Type-II):", FN1)
print("Type-I Error Rate in %:", type1_lr * 100)
print("Type-II Error Rate in %:", type2_lr * 100)
```

### 2. Decision Tree

```python
confusion_matrix_dt = confusion_matrix(Y_test, pred_dt)
print("Decision Tree Confusion Matrix:")
print(confusion_matrix_dt)

TN2, FP2, FN2, TP2 = confusion_matrix_dt.ravel()
type1_dt = FP2 / (FP2 + TN2)
type2_dt = FN2 / (FN2 + TP2)

print("FP (Type-I):", FP2)
print("FN (Type-II):", FN2)
print("Type-I Error Rate in %:", type1_dt * 100)
print("Type-II Error Rate in %:", type2_dt * 100)
```

### 3. K Nearest Neighbor (KNN)

```python
confusion_matrix_knn = confusion_matrix(Y_test, pred_knn)
print("K Nearest Neighbor Confusion Matrix:")
print(confusion_matrix_knn)

TN3, FP3, FN3, TP3 = confusion_matrix_knn.ravel()
type1_knn = FP3 / (FP3 + TN3)
type2_knn = FN3 / (FN3 + TP3)

print("FP (Type-I):", FP3)
print("FN (Type-II):", FN3)
print("Type-I Error Rate in %:", type1_knn * 100)
print("Type-II Error Rate in %:", type2_knn * 100)
```

### 4. Naive Bayes

```python
confusion_matrix_nb = confusion_matrix(Y_test, pred_nb)
print("Naive Bayes Confusion Matrix:")
print(confusion_matrix_nb)

TN4, FP4, FN4, TP4 = confusion_matrix_nb.ravel()
type1_nb = FP4 / (FP4 + TN4)
type2_nb = FN4 / (FN4 + TP4)

print("FP (Type-I):", FP4)
print("FN (Type-II):", FN4)
print("Type-I Error Rate in %:", type1_nb * 100)
print("Type-II Error Rate in %:", type2_nb * 100)
```

### 5. Neural Network

```python
confusion_matrix_nn = confusion_matrix(Y_test, pred_nn)
print("Neural Network Confusion Matrix:")
print(confusion_matrix_nn)

TN5, FP5, FN5, TP5 = confusion_matrix_nn.ravel()
type1_nn = FP5 / (FP5 + TN5)
type2_nn = FN5 / (FN5 + TP5)

print("FP (Type-I):", FP5)
print("FN (Type-II):", FN5)
print("Type-I Error Rate in %:", type1_nn * 100)
print("Type-II Error Rate in %:", type2_nn * 100)
```
