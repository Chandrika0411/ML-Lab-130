## Experiment-7: Ensemble Learning for House Price Prediction.

```python
import time 
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
# Laod Dataset
data = fetch_california_housing()
print(data.keys())
```

```python
X = pd.DataFrame(data.data,columns=data.feature_names)
Y = data.target
    
print(X.head())
print("Samples:",len(X))
print(Y[:10])
```

```python
# Train-Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size=0.2, random_state =42)
```

###  Regression Tree

```python
tree=DecisionTreeRegressor()

start=time.perf_counter()
tree.fit(X_train,Y_train)
tree_time=time.perf_counter()-start

pred_tree_test=tree.predict(X_test)
error_tree_test=mean_absolute_error(Y_test,pred_tree_test)

pred_tree_train=tree.predict(X_train)
error_tree_train=mean_absolute_error(Y_train,pred_tree_train)

print("Regression Tree Train Error:",error_tree_train)
print("Regression Tree Test Error:", error_tree_test)
print("Regression Tree Time:", tree_time)
```

### Random Forest

```python
rf=RandomForestRegressor(n_estimators=100)

start=time.perf_counter()
rf.fit(X_train,Y_train)
rf_time=time.perf_counter()-start

pred_rf_test=rf.predict(X_test)
error_rf_test=mean_absolute_error(Y_test,pred_rf_test)

pred_rf_train=rf.predict(X_train)
error_rf_train=mean_absolute_error(Y_train,pred_rf_train)

print("Random Forest Train Error:",error_rf_train)
print("Random Forest Test Error:", error_rf_test)
print("Random Forest Time:", rf_time)
```

## Gradient Boost

```python
gb=GradientBoostingRegressor(n_estimators=300,learning_rate=0.5,max_depth=5)

start=time.perf_counter()
gb.fit(X_train,Y_train)
gb_time=time.perf_counter()-start

pred_gb_test=gb.predict(X_test)
error_gb_test=mean_absolute_error(Y_test,pred_gb_test)

pred_gb_train=gb.predict(X_train)
error_gb_train=mean_absolute_error(Y_train,pred_gb_train)

print("Gradient Boosting Train Error:",error_gb_train)
print("Gradient Boosting Test Error:", error_gb_test)
print("Gradient Boosting Time:", gb_time)
```

## Comparison of Algorithm

### 1. Train Error

```python
print("\n-------Comparison For Train-------")
print("Regression Tree Test:", error_tree_train)
print("Randon Forest Test:", error_rf_train)
print("Gradient Boosting Test:", error_gb_train)
```

### 2. Test Error

```python
print("\n-------Comparison For Test-------")
print("Regression Tree Test:", error_tree_test)
print("Randon Forest Test:", error_rf_test)
print("Gradient Boosting Test:", error_gb_test)
```

### 3. Time Error

```python
print("\n-------Comparison For Time-------")
print("Regression Tree Test:", tree_time)
print("Randon Forest Test:", rf_time)
print("Gradient Boosting Test:", gb_time)
```
